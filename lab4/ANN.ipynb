```python
from keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
data = pd.read_csv('C://Users/baljot/Desktop/csc177 notebooks/jupyter notebook/Admission_Predict_Ver1.1_small_data_set_for_Linear_Regression.csv', header = 'infer')
```

```python
def encode_text_dummy(df, name):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = "{}-{}".format(name, x)
        df[dummy_name] = dummies[x]
    df.drop(name, axis=1, inplace=True)
```

```python
# Encode text values to indexes(i.e. [1],[2],[3] for red,green,blue).
def encode_text_index(df, name):
    le = preprocessing.LabelEncoder()
    df[name] = le.fit_transform(df[name])
    return le.classes_

# Convert a Pandas dataframe to the x,y inputs that TensorFlow needs
import collections
def to_xy(df, target):
    result = []
    for x in df.columns:
        if x != target:
            result.append(x)
    # find out the type of the target column. 
    target_type = df[target].dtypes
    target_type = target_type[0] if isinstance(target_type, collections.Sequence) else target_type
    # Encode to int for classification, float otherwise. TensorFlow likes 32 bits.
    if target_type in (np.int64, np.int32):
        # Classification
        dummies = pd.get_dummies(df[target])
        return df[result].values.astype(np.float32), dummies.values.astype(np.float32)
    else:
        # Regression
        return df[result].values.astype(np.float32), df[target].values.astype(np.float32)
```

```python
data = data.drop('Serial No.', axis = 1)
data.head(10)
```

```python
admissionChance = []
targetVal = 0.70
for row in data['Chance of Admit ']:
    if row >= 0.70:
        admissionChance.append(1.0)
    elif row < 0.70:
        admissionChance.append(0.0)
data['Chance of Admit '] = admissionChance
Admissions = encode_text_index(data,'Chance of Admit ')
data.head(11)
```

```python
testData = []
for i,row in data.iterrows() :
    if(i<151):
        testData.append(row)
testData = pd.DataFrame(testData)
testData.head()
```

```python
encode_text_index(testData,'Chance of Admit ')
Admissions
```

```python
X,Y = to_xy(data,'Chance of Admit ')
testX, testY = to_xy(testData,'Chance of Admit ')
```

```python
print(X.shape)
print (Y.shape)
Y
```

```python
model = Sequential()
model.add(Dense(12, input_dim = X.shape[1], activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam')
model.fit(X,Y,verbose=2, epochs=120)
```

```python
pred = model.predict(testX)
print (pred[0])
```

```python
pred = np.argmax(pred,axis = 1)
```
