```python
import pandas as pd

data = pd.read_csv('imdb_dataset.csv')
```

```python
data.head()
```

```python
data = data.drop (['Unnamed: 0', 'title_type', 'thtr_rel_month', 'thtr_rel_day', 'best_dir_win', 
                   'dvd_rel_year', 'dvd_rel_month', 'dvd_rel_day', 'actor1', 'actor2', 'actor3', 
                   'actor4', 'actor5', 'imdb_url', 'rt_url', 'best_pic_nom', 
                   'best_actor_win', 'best_actress_win', 'critics_score', 'director', 
                    'imdb_num_votes', 'studio', 'runtime',
                   'audience_score', 'best_pic_win'], axis = 1)
```

```python
data.head()
```

```python
import numpy as np

#Data in the top200_box
data = data.replace('no', '0')
data = data.replace('yes', 1)

#Genre
data = data.replace('Drama', 0)
data = data.replace('Comedy', 1)
data = data.replace('Horror', 2)
data = data.replace('Documentary', 3)
data = data.replace('Action & Adventure', 4)
data = data.replace('Art House & International', 5)
data = data.replace('Musical & Performing Arts', 6)
data = data.replace('Mystery & Suspense', 7)
data = data.replace('Animation', 8)
data = data.replace('Science Fiction & Fantasy', 9)
data = data.replace('Other', 10)

#Audience rating
data = data.replace('Upright', 0)
data = data.replace('Spilled', 1)

#Critics rating
data = data.replace('Fresh', 0)
data = data.replace('Rotten', 1)
data = data.replace('Certified Fresh', 2)

#mpaa_rating
data = data.replace('Unrated', np.nan)
data = data.dropna()

#mpaa_rating
data = data.replace('R', 0)
data = data.replace('PG-13', 1)
data = data.replace('PG', 2)
data = data.replace('G', 3)
data = data.replace('NC-17', 4)

data.head(20)
```

```python
from sklearn import cluster

data2 = data.drop(['title'],axis=1)
k_means = cluster.KMeans(n_clusters=2, max_iter=50, random_state=1)
k_means.fit(data2) 
labels = k_means.labels_
pd.DataFrame(labels, index=data['title'], columns=['Cluster ID'])
```

```python
centroids = k_means.cluster_centers_
pd.DataFrame(centroids, columns=data2.columns)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

numClusters = [1, 2, 3, 4, 5, 6, 7]
SSE = []
for k in numClusters:
    k_means = cluster.KMeans(n_clusters=k)
    k_means.fit(data2)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```

Hierarchical Clustering

```python
#Single LIN(MN). ALL THE TITLES
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
%matplotlib inline

title = data['title']
Y = data['imdb_rating']

X = data.drop(['title','imdb_rating'],axis=1)
Z = hierarchy.linkage(X.as_matrix(), 'single')
dn = hierarchy.dendrogram(Z,labels=title.tolist(),orientation='right')
```

```python
#Full Data Set to find Average
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
%matplotlib inline

title = data['title']
Y = data['imdb_rating']
X = data.drop(['title','imdb_rating'],axis=1)
Z = hierarchy.linkage(X.as_matrix(), 'average')
dn = hierarchy.dendrogram(Z,labels=title.tolist(),orientation='right')
```
